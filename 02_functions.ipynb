# Q13
- write a python program (as function) to calculate the sum of squares of positive intergers from 1 to any given n.

```python
# Function to calculate sum of squares from 1 to n
def sum_of_squares(n):
    total = 0
    for i in range(1, n + 1):
        total += i ** 2  # Add square of i to total
    return total

# Example usage
n = int(input("Enter a positive integer n: "))

if n > 0:
    result = sum_of_squares(n)
    print(f"Sum of squares from 1 to {n} is {result}")
else:
    print("Please enter a positive integer.")
```

# Q14
- write a python function named "add_corresponding_numbers" that takes two lists of numbers as input & returns a nes list containing the sum of  corresponding elements from the both lists. if the two lists are not of equal n the fuction should display the message.

```python
def add_corresponding_numbers(list1, list2):
    # Check if lists are of equal length
    if len(list1) != len(list2):
        print("The two lists are not of equal length!")
        return None  # Return None when lists are unequal
    
    # Create a new list with sums of corresponding elements
    result = []
    for i in range(len(list1)):
        result.append(list1[i] + list2[i])
    
    return result

# Example usage
list_a = [1, 2, 3]
list_b = [4, 5, 6]

sums = add_corresponding_numbers(list_a, list_b)
if sums is not None:
    print("Sum of corresponding elements:", sums)
```

# Q15
- write a oython function named "calculate_total" that calculates & returns the total cost after applying a discount.the function should,
    1. take price as a required argument
    2. use quantity as an optional argument with the default value of quantity.
    3. use discount as an optional argument with the default valie = 0.
    4. apply the discount as a percentage on the total price.

```python
def calculate_total(price, quantity=1, discount=0):
    """
    Calculates the total cost after applying a percentage discount.
    price     → required
    quantity  → optional (default = 1)
    discount  → optional (default = 0)
    """

    # Calculate initial total
    total = price * quantity

    # Apply discount percentage
    total_after_discount = total - (total * (discount / 100))

    return total_after_discount


# Example usage
amount = calculate_total(100, quantity=3, discount=10)
print("Total after discount:", amount)
```

# Q16
- write a python function named "sum_numbers" that accepts any number of numeric arguments & returns their sum.

```python
def sum_numbers(*numbers):
    """
    Accepts any number of numeric arguments
    and returns their sum.
    """
    total = 0
    for num in numbers:
        total += num
    return total


# Example usage
result = sum_numbers(5, 10, 15, 20)
print("Sum:", result)
```

# Q17
- write python function name "average_positive" that accepts any number of numeric argumenr & calculate the average of all positive numbers only. if no positive numbers provided, the function should display "No positive numbers provided"

```python
def average_positive(*numbers):
    """
    Accepts any number of numeric arguments and
    calculates the average of only the positive numbers.
    Prints a message if no positive numbers are provided.
    """

    positives = []  # list to store positive numbers

    # Collect only positive numbers
    for num in numbers:
        if num > 0:
            positives.append(num)

    # Check if we have any positives
    if len(positives) == 0:
        print("No positive numbers provided")
        return None

    # Calculate average
    avg = sum(positives) / len(positives)
    return avg


# Example usage
result = average_positive(3, -5, 10, -2, 7)

if result is not None:
    print("Average of positive numbers:", result)
```

# Q18
- write a python function named "average_score" that:
    1. accept any number of keyword arguments:
        where:
            1. keys are subject names(Like maths, physics)
            2. values are scores obtains in each subjects.
    2. calculate and print the average score of all subjects.
    3. if no subjects are provided print "no subjects provided"

```python
def average_score(**scores):
    """
    Accepts any number of keyword arguments where:
        - keys are subject names
        - values are scores
    Calculates and prints the average score.
    If no subjects provided, prints a message.
    """

    # Check if any subjects were provided
    if len(scores) == 0:
        print("No subjects provided")
        return

    # Extract all score values
    total = sum(scores.values())
    average = total / len(scores)

    # Print the average score
    print(f"Average score: {average}")
    

# Example usage
average_score(maths=85, physics=90, chemistry=78)
```

# Q19
- write a python function named "fahrenheit_to_celcius" that accepts a list of a tempratures in fahrenheit. use the math function to convert each temprature to celcius using the following formular:
    celcius = (Fahrenheit - 32) * 5/9
- returns a new list containing the tempratures in celcius, rounded to two decimal places.

```python
def fahrenheit_to_celcius(fahrenheit_list):
    """
    Accepts a list of temperatures in Fahrenheit.
    Converts each temperature to Celsius using:
        Celsius = (Fahrenheit - 32) * 5/9
    Returns a new list with the Celsius values
    rounded to two decimal places.
    """

    celcius_list = []

    for f in fahrenheit_list:
        c = (f - 32) * 5 / 9               # Convert to Celsius
        c = round(c, 2)                    # Round to 2 decimals
        celcius_list.append(c)             # Add to result list

    return celcius_list


# Example usage
temps_f = [32, 68, 77, 100]
temps_c = fahrenheit_to_celcius(temps_f)
print("Temperatures in Celsius:", temps_c)
```
