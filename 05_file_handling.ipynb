- write a python code that reads a file called "input.txt" and writes only the even numbered lines, into a new file called "even_lines.txt". consider the first line as line num 1.

```python
# Open the input file in read mode
with open("input.txt", "r") as input_file:
    lines = input_file.readlines()  # Read all lines into a list

# Open the output file in write mode
with open("even_lines.txt", "w") as even_file:
    # Loop through the lines with index
    for index, line in enumerate(lines, start=1):  # Line numbers start at 1
        if index % 2 == 0:  # Check if the line number is even
            even_file.write(line)  # Write the even line to the new file
```
