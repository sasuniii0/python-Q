# Q25
- write a python class named Car with the following specifications
    1. A class attribute "wheels" that stores the number 4.
    2. and instance method "drive_to" that prints "driving to<destination> on <numberofwheels> wheels"
    3. another instance method "show_whl=eel_count" that prints "<This car runs on <numberofwheels> wheels>"

```python
class Car:
    # Class attribute
    wheels = 4

    # Instance method to drive to a destination
    def drive_to(self, destination):
        print(f"Driving to {destination} on {Car.wheels} wheels")

    # Instance method to show the wheel count
    def show_wheel_count(self):
        print(f"This car runs on {Car.wheels} wheels")


# Example usage
my_car = Car()
my_car.drive_to("New York")
my_car.show_wheel_count()
```

# Q26
- create python class name circle that represent a circle. your class should be able to:
    1. store the radious of the circle.
    2. calculate & turn the circumference of the circle.
    3. calculate & return the area or circle.
    4. include a method "display_info" that prints the radious, area & circumference of the circle in a clear readable format.

```python
import math

class Circle:
    # Constructor to store the radius
    def __init__(self, radius):
        self.radius = radius

    # Method to calculate circumference
    def circumference(self):
        return 2 * math.pi * self.radius

    # Method to calculate area
    def area(self):
        return math.pi * self.radius ** 2

    # Method to display circle information
    def display_info(self):
        print(f"Circle Information:")
        print(f"Radius       : {self.radius}")
        print(f"Circumference: {self.circumference():.2f}")
        print(f"Area         : {self.area():.2f}")


# Example usage
c = Circle(5)
c.display_info()
```

# Q27
- create a python class called StudentRecords, that demostrate strong encapsulation. your class must include,
    1. a private attribute grades that store a list of intergers(0-100)
    2. a method "add_grade" that:
        - accept the grade and add it to the list.
        - calls a private method "validate_grade" to verify that the grade is a valid interger within range.
        - a method "get_average" that:
            * calculate & returns the students average score.
            * use private method "calculate_average" internally.

```python
class StudentRecords:
    def __init__(self):
        # Private attribute to store grades
        self.__grades = []

    # Private method to validate grade
    def __validate_grade(self, grade):
        if not isinstance(grade, int):
            raise ValueError("Grade must be an integer.")
        if grade < 0 or grade > 100:
            raise ValueError("Grade must be between 0 and 100.")
        return True

    # Private method to calculate average
    def __calculate_average(self):
        if len(self.__grades) == 0:
            return 0
        return sum(self.__grades) / len(self.__grades)

    # Public method to add a grade
    def add_grade(self, grade):
        if self.__validate_grade(grade):
            self.__grades.append(grade)
            print(f"Grade {grade} added successfully.")

    # Public method to get the average
    def get_average(self):
        return self.__calculate_average()


# Example usage
student = StudentRecords()
student.add_grade(85)
student.add_grade(90)
student.add_grade(78)

print("Average grade:", student.get_average())
```
