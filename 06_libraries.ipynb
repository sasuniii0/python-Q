- extract all elements that are:
    1. greater than 10 OR,less than 5.
    2. AND not equal to 15

````python
import numpy as np

# Create a 2D NumPy array
arr = np.array([
    [12, 5, 7, 20],
    [3, 15, 8, 2],
    [10, 25, 6, 18],
    [4, 11, 30, 9]
])

# Apply boolean indexing:
# (arr > 10)  → selects elements greater than 10
# (arr < 5)   → selects elements less than 5
# |            → OR condition between the two comparisons
# (arr != 15) → excludes the value 15
# &            → AND condition to combine both rules
result = arr[((arr > 10) | (arr < 5)) & (arr != 15)]

# Print the extracted elements
print(result)
````

```
import pandas as pd

    "Product" : ["Laptop" , "Mouse" , "Keyboard" , "Monitor" , "Headset"],
    "UnitsSold" : [12,50,34,20,40],
    "Price" : [12000,1500,3500,25000,6000],
    "Discount" : [0.10 , 0.05 , 0.08 , 0.12, 0.07]
})
```
1. calculate the final price after discount for each product.
2. calculate total revenue for each product.
for above part 1 and part 2 add both results as new columns to the DataFrame.
3. filter and display only the products where total revenue is greater than 500000.
4. calculate the total revenue of all products combined.

````python
import pandas as pd

# Create the DataFrame
df = pd.DataFrame({
    "Product": ["Laptop", "Mouse", "Keyboard", "Monitor", "Headset"],
    "UnitsSold": [12, 50, 34, 20, 40],
    "Price": [12000, 1500, 3500, 25000, 6000],
    "Discount": [0.10, 0.05, 0.08, 0.12, 0.07]
})

# 1. Calculate final price after discount
# Formula: FinalPrice = Price - (Price * Discount)
df["FinalPrice"] = df["Price"] - (df["Price"] * df["Discount"])

# 2. Calculate total revenue for each product
# Formula: TotalRevenue = FinalPrice * UnitsSold
df["TotalRevenue"] = df["FinalPrice"] * df["UnitsSold"]

# Display DataFrame after adding new columns
print("Updated DataFrame:")
print(df)

# 3. Filter and display products where total revenue is greater than 500000
high_revenue_products = df[df["TotalRevenue"] > 500000]

print("\nProducts with Total Revenue > 500000:")
print(high_revenue_products)

# 4. Calculate total revenue of all products combined
total_revenue_all = df["TotalRevenue"].sum()

print("\nTotal Revenue of all products combined:")
print(total_revenue_all)
````
