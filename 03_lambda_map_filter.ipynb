# Q20
- use the map function with two lists to add their corresponding elements & print the resulting sheet.

```python
# Two example lists
list1 = [1, 2, 3, 4]
list2 = [10, 20, 30, 40]

# Use map() to add corresponding elements
result = list(map(lambda a, b: a + b, list1, list2))

# Print the resulting list
print("Result:", result)
```

# Q21
- use the filter function to extract all numbers from the list:
    list = [12,15,18,20,25,30,35]
    that are devicable by 3 but not by 5, and print the resulting list.

```python
numbers = [12, 15, 18, 20, 25, 30, 35]

# Filter numbers divisible by 3 but NOT divisible by 5
result = list(filter(lambda n: n % 3 == 0 and n % 5 != 0, numbers))

# Print the resulting list
print("Filtered numbers:", result)
```

# Q23
- use lambda function with the map function to square all numbers in a given list and print the resulting list.

```python
numbers = [1, 2, 3, 4, 5]

# Use lambda + map to square each number
result = list(map(lambda x: x ** 2, numbers))

# Print the resulting list
print("Squared numbers:", result)
```
